```python
#Dokubo Tamunosaki
#DE.2023/0081
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('school_dressing_dataset.csv')

# Optional: Show first few rows
#Print(df.head())
```

```python
# Bar chart: Number of students by Gender

plt.figure(figsize=(8,5))
df['Gender'].value_counts().plot(kind='bar', color='mediumseagreen')
plt.title('Number of Students by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Students')
plt.tight_layout()
plt.show()
```

Chart Interpretation: Number of Students by Gender

• 	Axes:

• 	Y-axis (Vertical): Displays the gender categories — typically "Male" and "Female".

• 	X-axis (Horizontal): Shows the number of students in each gender category.

• 	Bars:

• 	Each horizontal bar represents the count of students who identify as either Male or Female.

• 	The length of the bar corresponds to how many students fall into that category.

• 	The color  gives a pleasant visual contrast, making the bars easy to distinguish.


```python
plt.figure(figsize=(12,6))
sns.histplot(data=df, x='Color', hue='Gender', multiple='dodge', shrink=0.8, palette='Set2', stat='count')
plt.title('Distribution of Uniform Colors by Gender')
plt.xlabel('Uniform Color')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Histogram for Distribution of Uniform Color by Gender

- Most Popular Colors: Which uniform colors are most common overall.

- Gender Trends: Whether certain colors are favored more by one gender.

- For example, if the bar for "Red" is taller for females than males, it suggests more females wore red.

- Uniform Diversity: Whether the school has a wide or narrow range of uniform colors.




```python
# Pie chart: Proportion of Uniform Top types

plt.figure(figsize=(8,8))
df['Uniform_Top'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=140,
    cmap='Set3',
    textprops={'fontsize': 12}
)
plt.title('Proportion of Uniform Top Types')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

Pie Chart to Show Proportion tof Uniform Top Types

- Most Common Uniform Top: The largest slice shows which top style is most popular.

- Style Diversity: If the pie is evenly split, it suggests a variety of styles are worn. If one dominates, it may reflect a school standard or tradition.

- Quick Proportional Insight: Pie charts are ideal for understanding relative proportions at a glance.


```python
# Area chart: Number of students by Uniform Color

color_counts = df['Color'].value_counts().sort_index()
plt.figure(figsize=(10,6))
plt.fill_between(color_counts.index, color_counts.values, color='lightgreen', alpha=0.7)
plt.plot(color_counts.index, color_counts.values, marker='o', color='green')
plt.title('Area Chart of Students by Uniform Color')
plt.xlabel('Uniform Color')
plt.ylabel('Number of Students')
plt.tight_layout()
plt.show()
```

Area Chart for Number of Students Uniform Color

- Most Popular Colors: The tallest peaks in the chart show which colors were worn by the most students.

- Least Popular Colors: The lowest dips indicate colors with fewer students.

- Visual Flow: Though the data is categorical, the area chart gives a smooth, flowing look that helps visualize overall distribution.


```python
# Box plot: Age distribution by Gender

plt.figure(figsize=(8,6))
sns.boxplot(x='Gender', y='Color', data=df, palette='Set2')
plt.title('Box Plot of Uniform Color Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Uniform Color')
plt.tight_layout()
plt.show()
```

Box Plot for Age distribution by Gender

- It show boxes for each gender, with the uniform colors treated as if they were numeric values.

```python
# Line chart: Number of students by Uniform Color
plt.figure(figsize=(8,5))
df['Color'].value_counts().sort_index().plot(kind='line', marker='o', color='teal')
plt.title('Number of Students by Uniform Color')
plt.xlabel('Uniform Color')
plt.ylabel('Number of Students')
plt.tight_layout()
plt.show()
```

Line Chart for Number of Students by Uniform Color

- Peaks in the line indicate popular uniform colors — many students wore them.

- Valleys show less common colors — fewer students chose them.

- The chart helps you quickly spot which colors are most and least favored.


```python
# Stacked bar chart: Number of students by Uniform Color and Gender

color_gender_counts = df.groupby(['Color', 'Gender']).size().unstack(fill_value=0)
color_gender_counts.plot(kind='bar', stacked=True, figsize=(10,6), colormap='Set2')
plt.title('Number of Students by Uniform Color and Gender (Stacked)')
plt.xlabel('Uniform Color')
plt.ylabel('Number of Students')
plt.tight_layout()
plt.show()
```

Stacked Bar Chart for Number of Students By Unifrom Color and Gender

- Tall Bars:
Indicate that a uniform color is popular overall.

- Segment Size:
Shows how many students of each gender wore that color.

- Gender Preference:
If one segment dominates a bar, it means that gender prefers that color more.


```python
# Heatmap: Number of students by Uniform Color and Gender

pivot = df.pivot_table(index='Color', columns='Gender', values='Student_ID', aggfunc='count', fill_value=0)
plt.figure(figsize=(8,6))
sns.heatmap(pivot, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Heatmap of Students by Uniform Color and Gender')
plt.xlabel('Gender')
plt.ylabel('Uniform Color')
plt.tight_layout()
plt.show()
```

Heatmap of Students by Uniform Color and Gender

- You can quickly compare which uniform colors are more popular among males vs. females.

- You can spot trends like:

- A color that’s mostly worn by one gender.

- A color that’s equally popular across genders.

- Colors that are rarely worn.


```python
# Horizontal Bar Chart
# Count occurrences of each Day_Occasion
occasion_counts = df['Day_Occasion'].value_counts()

# Plot horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(occasion_counts.index, occasion_counts.values, color='skyblue')
plt.xlabel('Number of Students')
plt.title('Student Count per Day Occasion')
plt.gca().invert_yaxis()  # Optional: highest count at the top
plt.tight_layout()
plt.show()
```

Horizontal Bar Chart for Student Count per Occasion

- Which school occasions had the highest or lowest student turnout.

- How student participation varies across different types of days.

- Helps identify the most engaging or popular events.
